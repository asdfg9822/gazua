# Instacart Market Basket Analysis

### Input File
* input/departments.csv
* input/orders.csv
* input/products.csv
* input/order_products__prior.csv
* input/order_products__train.csv


```python
import pandas as pd
```

## Load Dataset

```python
aisle = pd.read_csv("input/aisles.csv")
print(aisle.shape)
print(aisle.head())

departments = pd.read_csv("input/departments.csv")
print(departments.shape)
print(departments.head())

orders = pd.read_csv("input/orders.csv")
print(orders.shape)
print(orders.head())

products = pd.read_csv("input/products.csv")
print(products.shape)
print(products.head())

order_products__prior = pd.read_csv("input/order_products__prior.csv")
print(order_products__prior.shape)
print(order_products__prior.head())

order_products__train = pd.read_csv("input/order_products__train.csv")
print(order_products__train.shape)
print(order_products__train.head())
```

```python
ord_orderProductsPrior = pd.merge(orders, order_products__prior, on=['order_id'])
ord_orderProductsPrior.head()
```

```python
ord_orderProductsPrior_products = pd.merge(ord_orderProductsPrior, products, on=['product_id'])
```

```python
cProductName_cOrderHourOfDay = ord_orderProductsPrior_products[["product_name", "order_hour_of_day"]]
cProductName_cOrderHourOfDay
```

```python
print(type(cProductName_cOrderHourOfDay))
valueCounts_orderHourOfDayProduct = cProductName_cOrderHourOfDay.groupby(["order_hour_of_day"])["product_name"].value_counts()
print(valueCounts_orderHourOfDayProduct.shape)

valueCounts_orderHourOfDayProduct
```

```python
products_departments = pd.merge(products, departments, on=['department_id'])
print(products_departments.shape)
products_departments.head()
```
