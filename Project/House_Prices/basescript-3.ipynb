### Getting hard core
* Normality(정상상태) - 데이터가 정상 분포처럼 보여야 한다는 것. 
* Homoscedasticity(등분산성) - 예측 변수의 범위에 걸쳐 종속 변수가 동일한 수준의 편차를 보인다는 가정 (독립 변수의 모든 값에 걸쳐 동일한 오류 용어를 원하기 때문에 바람직하다?)
* Linearity(선형성) - scatter plot으로 검사하고 선형 패턴을 찾아보는 것. 만약 패턴이 선형적이지 않다면 데이터를 관찰할 필요가 있을 것.
* Absence of correlated errors(상관오류의 부재) - 상관오류란 한 오류가 다른 오류와 상호 연관될 때 발생. 예를 들어, 한가지 긍정적인 실수가 체계적으로 부정적인 오류를 발생시킨다면, 이는 이 변수들 사이에 관계가 있다는 것을 의미. 이러한 현상은 일부 패턴이 시간과 관련된 경우에 주로 발생한다.


```python
from scipy import stats

#histogram and normal probability plot
sns.distplot(df_train['SalePrice']);
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'], plot=plt)

# Histogram - Kurtosis and skewness.
# Normal probability plot - Data distribution should closely follow the diagonal that represents the normal distribution. 
```
